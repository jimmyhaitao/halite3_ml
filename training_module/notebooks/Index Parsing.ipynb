```python
import pickle
import gzip

# load the replay index
with gzip.open('/Users/Peace/Desktop/replays/INDEX.pkl', 'rb') as infile:
    master_index = pickle.load(infile)
```

```python
rnames = list(master_index.keys())
```

```python
master_index[rnames[0]]['game_statistics']
```

```python
good_replays = []
for rname in rnames:
    keep = False
    
    # Only focus on top player for now
    for player in master_index[rname]['players']:
        if 'Rachol ' in player['name']:
            keep = True
    if not keep:
        continue
        
    # Remove games where not everyone survived until the end
    num_turns = master_index[rname]['game_statistics']['number_turns']
    for player in master_index[rname]['game_statistics']['player_statistics']:
        if player['last_turn_alive'] != num_turns:
            keep = False
    
    if not keep:
        continue
        
    # Only take games where the player of interest won
        
    if keep:
        good_replays.append(rname)
```
